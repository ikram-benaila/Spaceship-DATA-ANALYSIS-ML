```python
def hotencode(dataframe,colonne):
##générer un one hot encode de cette colonne
  ohe = pd.get_dummies(dataframe[colonne],prefix=colonne)
  ohe.head()
  dataframe = dataframe.join(ohe)
  return dataframe
```

```python
#machine learning
y=dataframe['Transported'].values
x=dataframe.drop('Transported',axis=1).values
```

```python
# Standardisation des données
# Standardiser les données est important pour que :
# le modèle ne se focalise pas uniquement sur les colonnes avec de grandes valeurs
# toutes les valeurs soient prises en compte de manière équitable pour la prédiction
from sklearn.preprocessing import StandardScaler
SC=StandardScaler()
SC.fit(x)
x_scaled=SC.transform(x)
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
#pour voir la difference
print(x[0])
print(x_scaled[0])
```

```python
#model
#moindre carre (deminuer la valeur )
#il faut tester tout les models pour choissir le mellieur(%plus elves)
from sklearn.linear_model import LogisticRegression as LR
classifier=LR(random_state=42).fit(Xtrain,ytrain)
```

```python
y_predict=classifier.predict(Xtrain)
```

```python
#la prediction contient des valeurs juste ou fausse
y_predict
```

```python
#on a deux methode pour connaitre si le model a bien predire ou non
def accuracy(y,y_predict):
    acc=0 #nbr d'echentill correcte
    for index in range (len(y)):
        if y[index]==y_predict[index]:
            acc+=1
    return acc/len(y)*100 #pourcentage d'echentill correct predict
train_accuracy=accuracy(ytrain,y_predict)
print('train accuracy:{:2f}%'.format(train_accuracy))
```

```python
#on va faire le teste sur l'echentillon test
y_test_predict=classifier.predict(Xtest)
test_accuracy=accuracy(ytest,y_test_predict)
print(test_accuracy)
```
